```python
import pyautogui
import time

print("Mouse mover started. Press Ctrl+C to stop.")

while True:
    # Get current mouse position
    x, y = pyautogui.position()

    # Move mouse slightly and return to original spot
    pyautogui.moveTo(x + 1, y + 1, duration=0.1)
    pyautogui.moveTo(x, y, duration=0.1)

    # Wait 4 minutes before next move
    time.sleep(240)

```

```python
import ctypes
import time

# Windows virtual-key code for Shift key
VK_SHIFT = 0x10

def press_key(vk_code):
    ctypes.windll.user32.keybd_event(vk_code, 0, 0, 0)
    ctypes.windll.user32.keybd_event(vk_code, 0, 2, 0)

print("Windows auto key presser started. Press Ctrl+C to stop.")

try:
    while True:
        press_key(VK_SHIFT)
        print("Pressed Shift key.")
        time.sleep(240)  # every 4 minutes
except KeyboardInterrupt:
    print("Stopped.")
```
